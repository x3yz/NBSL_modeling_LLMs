```python
import numpy as np
import matplotlib.pyplot as plt
import networkx as nx

EPS = 0.0011
def normalize(p, eps = EPS):
    p = np.array(p, dtype=float) + eps
    p = p / p.sum()
    return p

def normalize_big(p, eps = EPS):
    p = np.array(p, dtype=float) + eps
    p = p / p.sum(axis=1, keepdims=True)
    return p


class RandomPool:
    """Centralized random pool with reproducible draws."""
    def __init__(self, seed=42):
        self.rng = np.random.RandomState(seed)
        self.choice_pool = self.rng.choice([0, 1, 2], size=1000, p=[0.25, 0.55, 0.2])
        self.ptr = 0

    def next_choice(self):
        choice = self.choice_pool[self.ptr]
        self.ptr += 1
        return choice

    def permutation(self, n):
        return self.rng.permutation(n)

    def reset(self):
        """Reset pointer for reproducibility across runs."""
        self.ptr = 0


class NBSL_agent:
    def __init__(self, idx, decision):
        self.idx = idx
        self.decision = decision


class Simulation:
    def __init__(self, N, top_n, movie_group, movietags, shuffled_tags, id_likelihoods, edges, decision_type, seed=42):
        self.N = N
        self.top_n = top_n
        self.movie_group = movie_group
        self.movietags = movietags
        self.id_likelihoods = id_likelihoods
        self.edges = edges
        self.T_max = 25  # fixed time horizon
        
        self.cur_id = [0] * N
        self.random_pool = RandomPool(seed)

        # create agents with infection status determined by infected_nodes
        self.agents = [NBSL_agent(i, decision=decision_type[i]) for i in range(N)]
        
        self.shuffled_movietags = []
        for id_movie, tags in enumerate(movietags):
            perm = self.random_pool.permutation(len(tags))
            shuffled = [tags[i] for i in perm]
            self.shuffled_movietags.append(shuffled)
        
        self.random_pool = RandomPool(seed)
        perm = self.random_pool.permutation(len(shuffled_tags))
        self.shuffled_newtags = [shuffled_tags[i] for i in perm]
        self.cur_split_id = 0  # counter for split tag usage

    def reset_randomness(self):
        """Reset the random pool for reproducibility."""
        self.random_pool.reset()
        self.random_pool = RandomPool(seed=42)
        self.cur_split_id = 0  # counter for split tag usage
        self.cur_id = [0] * self.N

        
    def get_tagid_with_likelihood(self, real_movie, agent, t, num_samples):
        behavior = self.random_pool.next_choice()
        tag_id = -1
        if (behavior == 0):
            idx = self.cur_split_id % len(self.shuffled_newtags)
            self.cur_split_id += 1
            tag_id, _ = self.shuffled_newtags[idx]
        elif (behavior == 1):
            tag_id = self.movie_group[real_movie][agent.idx][t]
        
        elif (behavior == 2):
            if (agent.decision == 0):
                # we either take shuffled (if next choice is 0) or normal (if next choice is 1)
                # or real_movie + 1 (if next choice is 2)
                id_movie = (real_movie + 1) % self.top_n
            else:
                id_movie = (real_movie + 2) % self.top_n
        
        if (tag_id == -1):
            id_tag_given = self.cur_id[id_movie] % len(self.shuffled_movietags[id_movie])
            self.cur_id[id_movie] += 1
            tag_id, _ = self.shuffled_movietags[id_movie][id_tag_given]

        print(f"choose {tag_id}... t={t}, {self.cur_split_id};")
        return tag_id, self.id_likelihoods[:num_samples, tag_id]

    def non_bayesian_social_learning(self, real_movie, num_samples):
        self.reset_randomness()
        # we assume that all the real movies group are of ~ the same size
        shuffled_times = self.random_pool.permutation(len(self.movie_group[real_movie][0]))
        shuffled_times = shuffled_times[ :self.T_max]

        ans = np.zeros((num_samples, self.T_max, self.N, 3))
        clueids = np.zeros((self.T_max, self.N))
        likelihoodBig = np.zeros((self.T_max, self.N, num_samples, self.top_n))

        # gather clues and likelihoods
        for id_t, t in enumerate(shuffled_times):
            for k, agent in enumerate(self.agents):
                clueids[id_t][k], likelihoodBig[id_t][k] = self.get_tagid_with_likelihood(real_movie, agent, t, num_samples)
        # belief updates across samples
        for sample in range(num_samples):
            prev_guess = [np.array([1/3, 1/3, 1/3]) for _ in range(self.N)]
            for id_t in range(self.T_max):
                for k in range(self.N):
                    likelihood = likelihoodBig[id_t][k][sample]
                    # aggregate neighbors' beliefs
                    neighbor_beliefs = [prev_guess[l] for l in self.edges[k]]
                    if neighbor_beliefs:
                        product = np.prod(neighbor_beliefs, axis=0)
                        product = product ** (1 / len(neighbor_beliefs))
                    else:
                        product = prev_guess[k]

                    ans[sample][id_t][k] = normalize(normalize(product) * normalize(likelihood))
                prev_guess = ans[sample][id_t]

        return ans
#     def bayesian_update(self, movie, agent, num_samples):
#         shuffled_times = self.random_pool.permutation(len(self.movie_group[movie][agent.idx]))
#         shuffled_times = shuffled_times[: self.T_max]

#         ans = np.zeros((num_samples, self.T_max, 3))
#         prev = np.tile(np.array([1/3, 1/3, 1/3]), (num_samples, 1))

#         for id_t, t in enumerate(shuffled_times):
#             tag_id, likelihood = self.get_tagid_with_likelihood(movie, agent, t, num_samples)
#             ans[:, id_t] = normalize_big(prev * normalize_big(likelihood))
#             prev = ans[:, id_t]

#         return ans
    
    
#     def global_identifiability(self, real_movie, normed_likelihoods, num_samples):
#         def divergence(p, q):
#             ans = 0.0
#             for i in range(p.shape[0]): # the whol T_max
#                 ans += p[i] * np.log(p[i] / q[i])
#             return ans
#         # we assume that all the real movies group are of ~ the same size
#         shuffled_times, clueids = self.get_clueids(real_movie, num_samples)
        
#         likelihoods = np.zeros((self.T_max, self.N, self.top_n))
        
#         for id_t, t in enumerate(shuffled_times):
#             for k, agent in enumerate(self.agents):
#                 print(id_t, k, clueids[id_t][k])
#                 likelihoods[id_t][k] = normed_likelihoods[clueids[id_t][k]]
#         dvs = []
#         for m in range(self.top_n):
#             if (m == real_movie):
#                 continue
#             dvs.append([])
#             for k in range(self.N):
#                 dv = divergence(likelihoods[:, k, m], likelihoods[:, k, real_movie])
#                 dvs[-1].append(dv)
#         for m in range(len(dvs)):
#             for k in range(self.N):
#                 print(dvs[m][k], end=", ")
#             print()
    def get_clueids(self, real_movie, num_samples):
        self.reset_randomness()
        # we assume that all the real movies group are of ~ the same size
        shuffled_times = self.random_pool.permutation(len(self.movie_group[real_movie][0]))
        shuffled_times = shuffled_times[ :self.T_max]

        ans = np.zeros((num_samples, self.T_max, self.N, 3))
        clueids = np.zeros((self.T_max, self.N), dtype=int)

        # gather clues and likelihoods
        for id_t, t in enumerate(shuffled_times):
            for k, agent in enumerate(self.agents):
                clueids[id_t][k], _ = self.get_tagid_with_likelihood(real_movie, agent, t, num_samples)
        return shuffled_times, clueids
```

Actual simulation run: 

```python
N = 23
top_n = 3
num_samples_done = 12

np.random.seed(42)
decision_type = np.random.choice([0, 1], size=(N,), p=[0.5, 0.5])

NUM_DONE_LIKELIHOODS = 140 # number of additional likelihoods computed

sim = Simulation(N, top_n, movie_group, movietags, sampled_split_tags[:NUM_DONE_LIKELIHOODS], id_likelihoods, edges, decision_type=decision_type, seed=42)

nbsl_ans = []
for real_movie in range(1):
    sim.reset_randomness()
    nbsl_ans = sim.non_bayesian_social_learning(real_movie, num_samples_done)
print("NBSL shape:", nbsl_ans.shape)
```
